```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Dell\Downloads\Jay Tatiya - Automobile_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    print('{} :{}'.format(col,df[col].unique))
```

```python
for col in df.columns:
    df[col].replace({'?':np.nan},inplace=True)
df.head()
```

```python
df.isnull().sum()
```

Missing values are found 


```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

there are many missing values in normalized_losses while other columns have fewer missing values

```python
num_col = ['normalized_losses', 'bore',  'stroke', 'horsepower', 'peak_rpm','price']
for col in num_col:
    df[col]=pd.to_numeric(df[col])
    df[col].fillna(df[col].mean(), inplace=True)
df.head()
```

here the Nan values are replaced by the mean

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='inferno')
```

+ve correlation 
Price – wheel_base, length, width, curb_weight, engine_size, bore, horsepower
wheelbase – length, width, height, curb_weight, engine_size, price
horsepower – length, width, curb_weight, engine_size, bore, price

-ve correlation
price-highway_mpg, city_mpg
city – wheel base, length, width, curb_weight, engine_size, bore, horsepower, price
